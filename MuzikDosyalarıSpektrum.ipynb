# FREQUENCY SPECTRUMS OF CLASSICAL MUSICS

## CLASSIC 1:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import read
```

```python
soundFile = 'Voices/1Klasik.wav'
Fs, x = read(soundFile)
%matplotlib notebook

print(x.shape) # x'in boyut bilgisi
print(Fs)
xn = x[:,1] # x'in sadece ilk sütunundaki örnekler alınmıştır (1. kanalın örnekleri)

N = len(x)
n=np.arange(0,N) # örnekleme indisinin 0'dan iki periyot olacak şekilde array olarak tanımlanması

M = 4 # Seyrek örnekleme (down sampling) oranı
xn_d = xn[np.arange(0, np.size(xn, 0), M)] # x[n] işaretinden sadece M katlarındaki örneklerin alınması
N_d = len(xn_d)
n_d = np.arange(0,N_d) # x_d[n] işaretinin indis dizisi
from IPython.lib.display import Audio
print('x[n] işareti için:')
display(Audio(xn, rate=Fs))
print('x_d[n] işareti için:')
display(Audio(xn_d, rate=Fs/M))

# klasik sinyallinin Fourier Transformu
w1 = np.arange(-np.pi, np.pi, 2*np.pi/N_d) # omega ekseninin -pi ile +pi arasında tanımlanması
xw1 = np.fft.fftshift(np.fft.fft(xn_d,N_d)/N_d) # ayrık zamanlı işaretin Fourier transformu

print(len(w1))
print(len(xw1))

plt.figure()
plt.stem(w1*Fs/(2*np.pi*M),abs(xw1)) # X(w) işaretinin grafiğinin çizdirilmesi

plt.xlabel("$Hz$") # grafiğin x ekseninin isimlendirilmesi
plt.ylabel("$|X(\omega)|$") # grafiğin y ekseninin isimlendirilmesi
plt.title("klasik1 $|X(\omega)|$") # grafiğin y ekseninin isimlendirilmesi

#Max Noktalarının Tespiti
print("Klasik 1")
xmax = abs(xw1).max()
print("Max Y Noktası:")
print(xmax)
y = w1*Fs/(2*np.pi*M)
ymax = y[np.argmax(abs(xw1))]
print("Max X Noktası:")
print(ymax)
```
