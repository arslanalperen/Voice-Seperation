# KONUSMA DOSYALARI SPEKTRUM

```python
import matplotlib.pyplot as plt # grafik çizimi için gerekli kütüphanenin aktifleştirilmesi
import numpy as np
from scipy.io.wavfile import read # Ses dosyasının okunması için gerekli kütüphane
%matplotlib notebook
```

## KONUSMA 1:

```python
Fs, x = read('Voices/B14_people01.wav')

print(x.shape)
print(Fs)

xn = x[:,1]

N = len(x)
n = np.arange(0,N)

w = np.arange(-np.pi, np.pi, 2*np.pi/N)
xw = np.fft.fftshift(np.fft.fft(xn,N)/N)

plt.figure()
plt.stem(w[110250:115250]*Fs/(2*np.pi),abs(xw[110250:115250]))
plt.xlabel("$Hz$")
plt.ylabel("$|X(\omega)|$")
plt.title("konusma1 $|X(\omega)|$")

#Max Noktalarının Tespiti
print("Konusma 1")
xmax = abs(xw).max()
print("Max Y Noktası:")
print(xmax)
y = w*Fs/(2*np.pi)
ymax = y[np.argmax(abs(xw))]
print("Max X Noktası:")
print(ymax)
```
