# FREQUENCY SPECTRUMS OF CLASSICAL MUSICS

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import read
```

## CLASSIC 1:

```python
Fs, x = read('Voices/B14_classical01.wav')

print(x.shape)
print(Fs)

xn = x[:,1]

N = len(x)
n = np.arange(0,N)

w = np.arange(-np.pi, np.pi, 2*np.pi/N)
xw = np.fft.fftshift(np.fft.fft(xn,N)/N)

print(len(w))

plt.figure()
plt.stem(w[110250:115250]*Fs/(2*np.pi),abs(xw[110250:115250]))
plt.xlabel("$Hz$")
plt.ylabel("$|X(\omega)|$")
plt.title("klasik1 $|X(\omega)|$")

#Max NoktalarÄ±nÄ±n Tespiti
print("Klasik1 1")
xmax = abs(xw).max()
print("Max Y NoktasÄ±:")
print(xmax)
y = w*Fs/(2*np.pi)
ymax = y[np.argmax(abs(xw))]
print("Max X NoktasÄ±:")
print(ymax)
```
